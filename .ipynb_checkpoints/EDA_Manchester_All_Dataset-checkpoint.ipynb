## Linear Regression Data Modeling 

### Select the highly coorelated feature

```python
# 1. Calculate the correlation matrix
correlation_matrix = waterfront_properties.corr().abs() # Using absolute values for magnitude
```

```python
# Define the correlation threshold
threshold = 0.5
```

```python
# 2. Select highly correlated features with the 'Price' variable
correlations_with_target = correlation_matrix['price']
```

```python
# Filter for features where the absolute correlation with 'Price' is greater than the threshold
relevant_features_with_target = correlations_with_target[correlations_with_target > threshold].index.tolist()
```

```python
# Select the upper triangle of the correlation matrix to avoid duplicate pairs
upper_tri = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
```

```python
# Find columns to drop based on high correlation with other features
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]
```

```python
print("Features highly correlated with the 'price' variable (absolute correlation > 0.5):")
print(relevant_features_with_target)

print("\nColumns to consider dropping due to high inter-feature correlation (absolute correlation > 0.5):")
print(to_drop)
```

```python
# Selecing the columns with the high correlation
waterfront_properties = waterfront_properties[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'grade', 'living_area']]
```

```python
waterfront_properties.plot(
    kind='box', 
    subplots=True, 
    sharey=False, 
    figsize=(9, 5)
)
plt.subplots_adjust(wspace=0.9) 
plt.show()
```

### Importing necessary libraries

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, mean_absolute_error 
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Splitting variables
X = waterfront_properties.iloc[:, 1:]
y = waterfront_properties.iloc[:, :1] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Prediction result
y_pred_test = model.predict(X_test)     # predicted value of y_test
y_pred_train = model.predict(X_train)   # predicted value of y_train
```

### Model Performance Evaluation

```python
model.score(X_test,y_test)
```

```python
# Evaluate the model using different metrics
r2 = r2_score(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse) # Calculate RMSE manually

print(f"R-squared (R2) Score: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
